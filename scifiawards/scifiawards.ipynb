```python
import pandas as pd
```

```python
hugos = pd.read_csv('hugos.csv')
hugos['org'] = 'Hugos'
nebulas = pd.read_csv('nebulas.csv')
nebulas['org'] = 'Nebulas'
wfa = pd.read_csv('wfa.csv')
wfa['org'] = 'World Fantasy'
locus = pd.read_csv('locus.csv')
locus['org'] = 'Locus'

scifi = pd.concat([hugos, nebulas, wfa, locus])
scifi = scifi[['title', 'author', 'award', 'org', 'year', 'outcome', 'published']]
scifi = scifi.sort_values(['year', 'award', 'org'], ascending=[False, True, True]).reset_index(drop=True)
```
