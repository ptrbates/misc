```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
base_url = 'http://www.sfadb.com/Locus_Awards_'
url_list = [base_url + str(num) for num in reversed(range(1971, 2018))]
```

```python
def get_locus(links):
    print('.' * len(links))
    listoflists = []
    
    for link in links:
        time.sleep(1)
        r = requests.get(link)
        text = r.text
        soup = BeautifulSoup(text, 'html.parser')

        year = soup.title.get_text()[-4:]

        categories = [category.get_text() for category in soup.find_all('div', {'class': 'category'})]
        awards_of_interest = ['Sf Novel', 'Fantasy Novel', 'Novella', 'Novelette', 'Short Story', 'Novel', 'Short Fiction']
        award_index = [categories.index(award) for award in awards_of_interest if award in categories]
        
        awards_this_year = [award for award in awards_of_interest if award in categories]
        awards = [[item.get_text() for item in link.find_all('li')] for link in soup.find_all('ol')]
        awards = list(awards[i] for i in award_index)
        
        try:
            pubs = [[item.rsplit('(', maxsplit=1)[1].replace(')', '') for item in award] for award in awards]
        except:
            pubs = []
            for index, award in enumerate(awards):
                pubs.append([])
                for item in award:
                    try:
                        pubs[index].append(item.rsplit('(', maxsplit=1)[1].replace(')', ''))
                    except:
                        pubs[index].append('no publishing information available')                    
        
        inter = [[item.rsplit('(', maxsplit=1)[0] for item in award] for award in awards]
        authors = [[item.rsplit(',')[1] for item in inter2] for inter2 in inter]
        titles = [[item.rsplit(',')[0] for item in inter2] for inter2 in inter]

        for i in range(len(awards)):
            for index, value in enumerate(titles[i]):
                item = [year, awards_this_year[i], index + 1, titles[i][index], authors[i][index], pubs[i][index]]
                listoflists.append(item)

        print('.', end='')
    
    locus = pd.DataFrame(listoflists, columns=['year', 'award', 'outcome', 'title', 'author', 'published'])    
    return(locus)
```

```python
listoflists = []

locus = get_locus(url_list)
locus = locus.drop_duplicates()

locus.author = [author.lower().replace(r". ", r".").replace(r".", r". ").title() for author in locus.author]
locus.title = [title.replace('Winner: ', '').replace('(tie):', '') for title in locus.title]
locus.title = [title.replace('"', '').replace('”', '').replace('“', '').strip() for title in locus.title]
locus = locus.replace(to_replace='Sf Novel', value='Best SF Novel')
locus = locus.replace(to_replace='Fantasy Novel', value='Best Fantasy Novel')
locus = locus.replace(to_replace='Novella', value='Best Novella')
locus = locus.replace(to_replace='Novelette', value='Best Novelette')
locus = locus.replace(to_replace='Short Story', value='Best Short Story')
locus = locus.replace(to_replace='Novel', value='Best Novel')
locus = locus.replace(to_replace='Short Fiction', value='Best Short Fiction')
    
# Make the year column numeric
locus['year'] = pd.to_numeric(locus['year'])

# Export the dataframe to csv
filename = 'locus.csv'
locus.to_csv(filename, index=False)
```
