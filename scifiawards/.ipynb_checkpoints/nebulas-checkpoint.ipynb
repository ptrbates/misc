```python
# Import required libraries
import requests
from bs4 import BeautifulSoup
import re
import time
import pandas as pd
```

```python
url = 'https://nebulas.sfwa.org/awards-by-year/'

r = requests.get(url)
text = r.text

soup = BeautifulSoup(text, 'html.parser')
```

```python
# Retrieve a list of links to the award history pages
awardpattern = re.compile('https://nebulas.sfwa.org/award-year')

links = []
for link in soup.find_all('a'):
    if bool(awardpattern.search(link.get('href'))):
        links.append(link.get('href'))

links = sorted(list(set(links)), reverse=True)
```

```python
def get_nebulas(links):
    print('.' * len(links))
    listoflists = []

    for link in links:
        time.sleep(1)
        r = requests.get(link)
        text = r.text
        soup = BeautifulSoup(text, 'html.parser')
        year = soup.find_all('title')[0].get_text()[0:4]
    
        for award in ['Best Novel', 'Best Novella', 'Best Novelette', 'Best Short Story']:
            if award == 'Best Novel':
                stories = soup.find_all('ul', {'class': 'best-novel award_group'})[0].find_all('li')
            elif award == 'Best Novella':
                stories = soup.find_all('ul', {'class': 'best-novella award_group'})[0].find_all('li')
            elif award == 'Best Novelette':
                stories = soup.find_all('ul', {'class': 'best-novelette award_group'})[0].find_all('li')
            elif award == 'Best Short Story':
                stories = soup.find_all('ul', {'class': 'best-short-story award_group'})[0].find_all('li')
            for item in stories:
                if bool(item.find_all('i')):
                    outcome = "Winner"
                else:
                    outcome = "Finalist"
                title = item.find_all('a')[0].get_text()
                author = item.find_all('a')[1].get_text()
                try:
                    pubs = item.find_all('a')[2].get_text()
                except:
                    pubs = "no publishing information available"

                listoflists.append([year, award, outcome, title, author, pubs])
        print('.', end='')
    
    nebulas = pd.DataFrame(listoflists, columns=['year', 'award', 'outcome', 'title', 'author', 'published']) 
    return(nebulas)
```

```python
nebulas = get_nebulas(links)

nebulas = nebulas.drop_duplicates()

nebulas.author = [author.lower().replace(r". ", r".").replace(r".", r". ").title() for author in nebulas.author]
nebulas.title = [title[0:-1] if title[-1] == ',' else title for title in nebulas.title ]

# Make the year column numeric
nebulas['year'] = pd.to_numeric(nebulas['year'])

# Export the dataframe to csv
filename = 'nebulas.csv'
nebulas.to_csv(filename, index=False)
```
