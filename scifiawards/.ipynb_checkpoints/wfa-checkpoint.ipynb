```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
url = 'http://www.worldfantasy.org/awards/nominees/'

r = requests.get(url)
text = r.text
soup = BeautifulSoup(text, 'html.parser')

awards = soup.find_all('table', {'id': ''})[0]
header = ['First Name', 'Last Name', 'year', 'award', 'title', 'outcome']

long_list = [item.get_text() for item in awards.find_all('td')]
first_name = long_list[::6]
last_name = long_list[1::6]
year = long_list[2::6]
award = long_list[3::6]
title = long_list[4::6]
outcome = long_list[5::6]
```

```python
wfa = pd.DataFrame(np.transpose([first_name, last_name, year, award, title, outcome]), columns=header)
wfa.drop_duplicates()

wfa['author'] = wfa['First Name'] + ' ' + wfa['Last Name']
wfa.author = [author.lower().replace(r". ", r".").replace(r".", r". ").title() for author in wfa.author]
wfa.author = [' '.join(author.split()) for author in wfa.author]
wfa.title = [title.replace('"', '') for title in wfa.title]
wfa['year'] = pd.to_numeric(wfa['year'])
wfa = wfa.filter(items=['year', 'award', 'outcome', 'title', 'author'])
wfa = wfa.replace(to_replace='Novel', value='Best Novel')
wfa = wfa.replace(to_replace='Short Fiction', value='Best Short Fiction')
wfa = wfa.replace(to_replace='Long Fiction', value='Best Long Fiction')
wfa = wfa[wfa.award.isin(['Best Novel', 'Best Short Fiction', 'Best Long Fiction'])]

wfa.to_csv('wfa.csv', index=False)
```
